```python
import yfinance as yf
import pandas as pd
import talib as ta
import numpy as np
import openpyxl

# ฟังก์ชันเพื่อดึงและประมวลผลข้อมูลหุ้น
def process_stock(stock):
    df = yf.Ticker(stock + '.BK')
    try:
        df = df.history(interval='1H', start='2022-02-01', end='2023-12-30')
        if df.empty:
            print(f"No data for {stock}")
            return None

        df['up_band'], df['mid_band'], df['low_band'] = ta.BBANDS(df['Close'], 
                                                                  timeperiod=55, nbdevup=2, 
                                                                  nbdevdn=2, matype=0)
        df = df[['Close', 'up_band', 'mid_band', 'low_band']]
        df.reset_index(inplace=True)
        df.rename(columns={'index': 'Datetime'}, inplace=True)
        df['date'] = df['Datetime'].dt.date
        df['time'] = df['Datetime'].dt.time
        df.drop(['Datetime'], axis=1, inplace=True)
        df.dropna(inplace=True)  # ลบแถวที่มีค่าว่าง
        return df
    except Exception as e:
        print(f"Error processing {stock}: {e}")
        return None


# อ่านไฟล์ Excel ที่มีรายชื่อหุ้น
excel_path = r'C:\Users\ming\Desktop\MyJob\DSM1.1.2\data\จัดอันดับหุ้น std.xlsx'  # แทนที่ด้วย path ของไฟล์ Excel ของคุณ
stock_df = pd.read_excel(excel_path)
stock_list = stock_df['Stock'].astype(str).tolist() 

# กำหนดพาธสำหรับบันทึกไฟล์ Excel ผลลัพธ์
output_excel_path = r'C:\Users\ming\Desktop\MyJob\DSM1.1.2\data\set50.xlsx'  # แทนที่ด้วยพาธที่คุณต้องการบันทึกไฟล์

# สร้างไฟล์ Excel ใหม่สำหรับผลลัพธ์
writer = pd.ExcelWriter(output_excel_path, engine='openpyxl')

for stock in stock_list:
    # เพิ่มการตรวจสอบเพื่อแน่ใจว่า stock เป็นสตริงและไม่ว่าง
    if isinstance(stock, str) and stock:
        df = process_stock(stock)
        if df is not None:
            df.to_excel(writer, sheet_name=stock, index=False)
    else:
        print(f"Invalid stock name: {stock}")

try:
    writer.close()  # ใช้ close แทน save
    print("Data saved to Excel file.")
except Exception as e:
    print(f"Error saving the file: {e}")
```
