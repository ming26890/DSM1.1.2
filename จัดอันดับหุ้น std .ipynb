```python
import yfinance as yf
import pandas as pd
import talib as ta
import numpy as np
import math
import matplotlib.pyplot as plt
from copy import deepcopy

# โหลดไฟล์ Excel
file_path = r'C:\Users\ming\Desktop\MyJob\DSM1.1.2\history\stock_data.xlsx'  # แทนที่ด้วยที่ตั้งของไฟล์ Excel ของคุณ

# อ่านข้อมูลจากทุกๆ worksheet ในไฟล์ Excel
all_data = pd.read_excel(file_path, sheet_name=None)

# สร้าง dictionary สำหรับเก็บชื่อหุ้นและค่า standard deviation ของมัน
stock_std = {}

# คำนวณค่า standard deviation ของราคาปิดสำหรับแต่ละหุ้น
for stock_name, data in all_data.items():
    if 'Close' in data.columns:
        stock_std[stock_name] = data['Close'].std()

# สร้าง DataFrame จาก dictionary และเรียงลำดับมัน
std_df = pd.DataFrame(list(stock_std.items()), columns=['Stock', 'Std']).sort_values(by='Std', ascending=False)

# แสดงผล DataFrame
print(std_df)

# ระบุที่ตั้งและชื่อไฟล์สำหรับไฟล์ Excel ที่จะบันทึก
output_file_path = r'C:\Users\ming\Desktop\MyJob\DSM1.1.2\history/จัดอันดับหุ้น std.xlsx'  # แทนที่ด้วยที่ตั้งที่คุณต้องการบันทึก

# บันทึก DataFrame เป็นไฟล์ Excel
std_df.to_excel(output_file_path, index=False)
```
